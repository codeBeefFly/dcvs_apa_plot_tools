```python

import cv2
import matplotlib.pyplot as plt


img_path = r"D:\D_Documents\DocsCollection\ROS_WORLD\DCVS_WORLD_BASIS\dcvs_underground_precise\dcvs_underground_precise"
img_name = "dcvs_underground_precise_rviz.png"
img = cv2.imread(img_path + "/" + img_name, 0) #直接读为灰度图像

ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

thresh6 = cv2.adaptiveThreshold(img,  255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY_INV,21,0 ) 
thresh7 = cv2.adaptiveThreshold(img,  255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,21,0 ) 


cv2.imwrite(img_path + "/" + "dcvs_underground_precise_rviz2.png", img)
cv2.imwrite(img_path + "/" + "dcvs_underground_precise_rviz2.png", thresh1)
cv2.imwrite(img_path + "/" + "dcvs_underground_precise_rviz3.png", thresh6)
cv2.imwrite(img_path + "/" + "dcvs_underground_precise_rviz4.png", thresh7)

titles = ["img", "BINARY", "BINARY_INV", "TRUNC", "TOZERO", "TOZERO_INV", "ADA1", "ADA2"]
images = [img,thresh1,thresh2,thresh3,thresh4,thresh5,thresh6,thresh7]
for i in range(8):
    plt.subplot(2,4,i+1),plt.imshow(images[i], "gray")
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
